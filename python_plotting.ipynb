```python
import numpy as np
import matplotlib.pyplot as plt
#create theta array
th = np.linspace(0, 2*np.pi, 100)

a = 5
b = 13
#coordinates for centre of the ellipse
ox = 0
oy = 0
#creating x and y arrays from theta
x = ox + a*np.cos(th)
y = oy + b*np.sin(th)

#plotting the ellipse
plt.scatter(x, y, color='green')
plt.plot(x, y, color='red')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Ellipse')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#parabola

a = 2
x = np.linspace(-10, 10, 100)
y = x**2/(4*a)

#plotting the parabola
plt.scatter(x, y, color='blue')
plt.plot(x, y, color='black')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Parabola')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#create theta array
th = np.linspace(0, 2*np.pi, 100)

a = -2
b = np.sqrt(2)
#coordinates for centre of the circle
ox = 0
oy = 0
#creating x and y arrays from theta
x = ox + a*(1/np.cos(th))
y = oy + b*np.tan(th)

#plotting the hyperbola
plt.scatter(x, y, color='purple')
plt.plot(x, y, color='black')
plt.axis('equal')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Hyperbola')
plt.grid()
plt.show()
```
